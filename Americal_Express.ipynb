```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python

df = df.drop(['name', 'customer_id','occupation_type'], axis='columns')
df.head()
```

```python
dummies1 = pd.get_dummies(df.gender)
```

```python
dummies1.head(5)
```

```python
dummies2= pd.get_dummies(df.owns_car)
dummies2.head(5)
```

```python
dummies3= pd.get_dummies(df.owns_house)
dummies3.head(5)
```

```python
data = pd.concat([df, dummies1, dummies2, dummies3], axis='columns')
```

```python
data
```

```python
data = data.drop(['gender', 'owns_car', 'owns_house'], axis='columns')
data
```

```python
target = data.credit_card_default
train_data = data.drop(['credit_card_default'], axis='columns')
train_data
```

```python
train_data.columns[train_data.isna().any()]
train_data.age = train_data.age.fillna(train_data.age.mean())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data, target, test_size=0.3)
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
lr = LogisticRegression()
lr.fit(X_train, Y_train)
lr.score(X_test, Y_test)
```
